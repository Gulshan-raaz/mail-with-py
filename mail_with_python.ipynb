```
import smtplib
from getpass import getpass
```

```
server = 'smtp.gmail.com'
port = 587

conn = smtplib.SMTP(server, port)
conn.ehlo()
conn.starttls()
mymailid = "gulshanraj7399@gmail.com"
mymailpassword = "********"
```

```
mailto = input('Enter mailid of the person: \n')
```

```
print("Fill The Below Form \n")
sub = input('Subject:  \n')
message = input("Type Your Message : \n")
```

```
try:
    conn.login(mymailid, mymailpassword)
    conn.sendmail(mymailid, mailto, 'Subject: ' + sub + '\n\n' + message)
    conn.quit()
    print()
    print('Your Message Send Success')
except:
    print()
    print('Enable LessSecures From google privacy ')
```
